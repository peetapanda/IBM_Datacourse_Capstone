## Demo: How to work with database

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
# Retriving data into Pandas
df = pd.read_sql_query("select * from master;", conn)
df.head()
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
```

```python
# your code goes here
df['ConvertedComp'].plot(kind='hist')
plt.title('ConvertedComp')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df['Age'].plot(kind='box')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df.plot(x='Age', y='WorkWeekHrs', kind='scatter')
plt.title('Work Week Hours Based on Age')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
df.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs', alpha=0.5, s='Age')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""
db_desire_df = pd.read_sql_query(QUERY,conn)
```

```python
db_desire_df['DatabaseDesireNextYear'].value_counts()
```

```python
top5_db = db_desire_df['DatabaseDesireNextYear'].value_counts().head()
top5_db.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
# Filtering Data
age30to35_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_df = age30to35_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create Graph
median_df.plot(kind='bar', stacked=True)
plt.ylabel('Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
age45to60_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
median_df = age45to60_df.groupby('Age')['ConvertedComp'].median()
median_df.plot(kind='line')
plt.title('ConvertedComp of Respondents Aged 45-60')
plt.ylabel('Converted Comp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df['MainBranch'].value_counts().plot(kind='barh')
plt.show()
```
